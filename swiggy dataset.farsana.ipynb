```python
import plotly.express as px
import plotly.graph_objects as go
import warnings
```

```python
fig = px.histogram(df, x ="Price", nbins = 50, title = "Price Distribution")
fig
```

```python
fig1 = px.histogram(df, x ="Avg ratings", nbins = 20, title = "Average Ratings Distribution")
fig1
```

```python
fig4 = px.histogram(df, x='Delivery time', nbins=30, title='Delivery Time Distribution')
fig4.show()
```

```python
fig2  = px.scatter(df, x = "Avg ratings", y = "Price", color = "City", title = "Price v Average Ratings")
fig2.show()
```

```python
top_cities = df['City'].value_counts().head(10).reset_index()
top_cities.columns = ['City', 'Count']
fig3 = px.bar(
    top_cities, 
    x='City', 
    y='Count', 
    title='Top 10 Cities with Most Restaurants', color = 'City',
    labels={'City': 'City', 'Count': 'Number of Restaurants'}
)
fig3.show()
```

```python
top_food_types = df['Food type'].value_counts().head(10).reset_index()
top_food_types.columns = ['Food type', 'Counts']
fig5 = px.bar(top_food_types, x='Food type', y='Counts', title='Top 10 Food Types', 
              labels={'Food type': 'Food Type', 'Counts': 'Count of food types'})
fig5.show()
```

```python
city_avg_delivery=df.groupby('City')['Delivery time'].mean().reset_index().sort_values(by='Delivery time',ascending=False)
fig6=px.bar(city_avg_delivery,x='City',y='Delivery time',title='Average Delivery time by City',labels={'City':'City','Delivery time':'Avg Delivery time(mins)'})
fig6
```

```python
city_avg_price=df.groupby('City')['Price'].mean().reset_index().sort_values(by='Price',ascending=False)
fig7=px.bar(city_avg_price,x='City',y='Price',title='Average Price of Orders by City',labels={'City':'City','Price':'Avg price(INR)'})
fig7
```

```python
city_name='Bangalore'
top_food_types=df[df['City']==city_name]['Food type'].value_counts().reset_index().head(5)
top_food_types.columns=['Food Type', 'Count']
fig8=px.bar(top_food_types,x='Food Type',y='Count',title=f'Top 5 Foodtypes in{city_name}')
fig8
```

```python
city_ratings=df[df['City']==city_name]
fig9=px.histogram(city_ratings,x='Avg ratings',nbins=100,title=f'Average Rating Distribution in{city_name}',labels={'Avg ratings': 'Average Ratings'})
fig9
```

```python
fig10=px.scatter(city_ratings,x='Price',y='Delivery time',title=f'Price vs Delivery Time in {city_name}',color='Restaurant',labels={'Price': 'Price (INR)', 'Delivery time': 'Delivery Time (mins)'})
fig10
```
