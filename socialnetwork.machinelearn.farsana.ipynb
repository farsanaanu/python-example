```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r'C:\Users\USER\Downloads\fwdclassificationalgorithmslogisticknnsvmnaivebay\Social_Network_Ads.csv')
```

```python
dataset
```

```python
x=dataset.iloc[:,:-1].values
```

```python
x
```

```python
y=dataset.iloc[:,-1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
print(x_train)
x_train.shape
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.preprocessing import  StandardScaler
sc= StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(x_train,y_train)
```

```python
print(clf.predict(sc.transform([[30,87000]])))
```

```python
y_pred=clf.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
#Making the Confusion Matrix
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy_score(y_test,y_pred)
```

```python
#Visualising the Training set results
from matplotlib.colors import ListedColormap
x_set,y_set=sc.inverse_transform(x_train),y_train
x1,x2=np.meshgrid(np.arange(start=x_set[:,0].min()-10,stop=x_set[:,0].max()+10,step=0.25),np.arange(start=x_set[:,1].min()-1000,stop=x_set[:,1].max()+1000,step=0.25))
plt.contourf(x1,x2,clf.predict(sc.transform(np.array([x1.ravel(),x2.ravel()]).T)).reshape(x1.shape),alpha=0.75,cmap=ListedColormap(('red', 'green')))
plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set==j,0],x_set[y_set==j,1],c= ListedColormap(('red', 'green'))(i),label=j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
