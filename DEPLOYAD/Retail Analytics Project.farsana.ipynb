# RFM Customer Segmentation

```python
from datetime import datetime
```

```python
Retail['PurchaseDate']=pd.to_datetime(Retail['Date'])
current_date=datetime.now()
```

### RFM Metrics

```python
rfm=Retail.groupby('Customer_ID').agg(
    Recency=('Date',lambda x:(current_date-x.max()).days),
    Frequency=('Date','count'),
    Monetary=('Amount','sum')
).reset_index()
print("RFM Metrics:\n",rfm)
```

### RFM Score

```python
rfm['Recencyscore']=pd.qcut(rfm['Recency'],5,labels=[5,4,3,2,1])
rfm['FrequencyScore']=pd.qcut(rfm['Frequency'],5,labels=[1,2,3,4,5])
rfm['Monetaryscore']=pd.qcut(rfm['Monetary'],5,labels=[1,2,3,4,5])
rfm['RFMScore']=rfm['Recencyscore'].astype(str)+rfm['FrequencyScore'].astype(str)+rfm['Monetaryscore'].astype(str)
```

### Customer Segment

```python
def segment_customer(row):
    if row['RFMScore']=='555':
        return 'Best Customers'
    elif row['RFMScore']=='111':
        return 'Lost Customers'
    elif row['RFMScore'].startswith('5'):
        return 'Loyal Customers'
    elif row['RFMScore'].endswith('1'):
        return 'At Risk'
    else:
        return 'Others'
```

```python
rfm['Segment']=rfm.apply(segment_customer,axis=1)
print("\nCustomer Segments:\n",rfm[['Customer_ID','Recency','Frequency','Monetary','RFMScore','Segment']])
```

### Putting the results well organised into a Pandas DataFrame

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

# Sales Forecasting

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
data={
    'Date':pd.date_range(start='2023-01-03',periods=180,freq='D'),
    'Total_Purchases':np.random.randint(100,300,size=180)
}
```

### Create Dataframe

```python
df=pd.DataFrame(data)
df.set_index('Date',inplace=True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Total_Purchases'],color='green')
plt.title('Daily Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
```

### Split the Data into Training and Test Sets

```python
train_size=int(len(df)*0.8)
train,test=df[:train_size],df[train_size:]
```

### Fit the ARIMA model on Training set

```python
model=ARIMA(train['Total_Purchases'],order=(5,1,0))
model_fit=model.fit()
```

### Forecast the Sales on Test Data

```python
forecast=model_fit.forecast(steps=len(test))
```

```python
forecast
```

### Plot Actual and Forecast Sales

```python
plt.figure(figsize=(10,6))
plt.plot(train.index,train['Total_Purchases'],label='Training Data')
plt.plot(test.index,test['Total_Purchases'],label='Test Data',color='gray')
plt.plot(test.index,forecast,label='Forecast Data',color='red')
plt.title('Sales Fprecasting with ARIMA')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

### Evaluate the model using Mean Absolute Error

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae=mean_absolute_error(test['Total_Purchases'],forecast)
print(f'Mean Absolute Error (MAE):{mae}')
```

### Predict Future Sales

```python
future_forecast=model_fit.forecast(steps=30)
future_dates=pd.date_range(start=df.index[-1]+pd.Timedelta(days=1),periods=30,freq='D')
```
