# import essential libraries for predicting and accuracy checking

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
hospital['Gender']=label_encoder.fit_transform(hospital['Gender'])
hospital=pd.get_dummies(hospital,columns=['Status'])
```

```python
hospital['Gender']
```

```python
hospital.head()
```

```python
scaler=StandardScaler()
hospital[['Age']]=scaler.fit_transform(hospital[['Age']])
```

```python
hospital['Age']
```

# define features(x) and predict(Y)

```python
x=hospital.drop(hospital[['FZ me','Name','City','State','Patient type','Custom','Department_Name','Bed','Feedback','Age Bucket','Date']],axis=1)
y=hospital['FZ me']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print(y_pred)
```

```python
print(classification_report)
```

```python
accuracy= accuracy_score(y_test,y_pred)
print(f'Accuracy:{accuracy}')
```

```python
confusion_metrix=confusion_matrix(y_test,y_pred)
print(confusion_metrix)
```

```python
classification_report=classification_report(y_test,y_pred)
print(classification_report)
```
