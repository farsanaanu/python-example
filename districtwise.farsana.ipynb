```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\USER\district_wise_details.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
fig=plt.figure(figsize=(12,6))
axes=fig.add_axes([0,0,1,1])
axes.plot(df['district'].head(),df['fatalities'].head())
axes.set_title('District vs fatalities')
axes.set_xlabel('district')
axes.set_ylabel('fatalities')
```

```python
fig,ax=plt.subplots(1,3,figsize=(12,6))
ax[0].plot(df['district'],df['no_of_camps'])
ax[0].set_title('District vs no_of_camps')
ax[0].set_xlabel('district')
ax[0].set_ylabel('no_of_camps')
ax[1].plot(df['district'],df['actual_rainfall_in_mm'],'r')
ax[1].set_title('district vs actual_rainfall_in_mm')
ax[1].set_xlabel('district')
ax[1].set_ylabel('actual_rainfall_in_mm')
ax[2].plot(df['district'],df['normal_rainfall_in_mm'],'g')
ax[2].set_title('district vs normal_rainfall_in_mm')
ax[2].set_xlabel('district')
ax[2].set_ylabel('normal_rainfall_in_mm')
ax[2].set_xlim('Thiruvananthapuram','Pathanamthitta')
ax[1].set_xlim('Thiruvananthapuram','Pathanamthitta')
ax[0].set_xlim('Thiruvananthapuram','Pathanamthitta')
plt.tight_layout()
```

```python
sns.barplot(x=df['district'].head(),y=df['no_of_landslides'].head()).set_title('district vs no_of_landslides ')
```

```python
sns.scatterplot(x=df['district'].head(),y=df['full_damaged_houses'].head()).set_title('District vs full_damaged_houses')
```

```python
fig=plt.figure(figsize=(14,8))
ax=fig.add_axes([0,0,1,1])
ax.pie(df['normal_rainfall_in_mm'])
ax.set_title('District vs normal_rainfall_in_mm')
plt.pie(df['normal_rainfall_in_mm'],labels=df['district'],autopct='%1.1f%%');
plt.legend(loc=4)
```

```python
sns.distplot(df['full_damaged_houses']).set_title('District vs full_damaged_houses');
```
